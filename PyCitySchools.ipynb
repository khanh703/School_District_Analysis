### Why are these ranges what they are given the numbers above

Seems arbitrary. Understand we don't want 578 as first bin because there is only one. With the std being 28, why does 578 become 585 and 641 become 675? Or the aim is to attempt to group evenly as possible?

585 - 578 = 7

615 - 591 = 24

645 - 628 = 17

675 - 641 = 34

```python
spending_bins = [0, 585, 630, 645, 675]
```

The cut() function places each value of the per_school_capita Series into one of four bins. The bins are created by a list of five numbers given as the spending_bins that define the edges.

```python
pd.cut(per_school_capita, spending_bins)
```

```python
per_school_capita.groupby(pd.cut(per_school_capita, spending_bins)).count()
```

## Rename bins

```python
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]
```

```python
# Categorize spending based on the bins.
per_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)

per_school_summary_df
```

Human decision making:

"Small" schools have fewer than 1,000 students

"Medium" schools have 1,000–1,999 students

"Large" schools have 2,000–5,000 students

### Create the series to cut() with

Use these 2 series to cut() the 'per_school_summary_df'

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

- cut into per_school_summary_df
- (param1) add new column "School Size"
- (param2) range
- (param3) labels

```python
per_school_summary_df["School Size"] = pd.cut(per_school_summary_df["Total Students"], size_bins, labels=group_names)
per_school_summary_df.head()
```
